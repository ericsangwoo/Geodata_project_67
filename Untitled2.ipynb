```python
import numpy as np
import matplotlib.pyplot as plt

# Given data from the slide
salinity_ppm = np.array([10000, 12000, 14000, 17000, 20000,
                         30000, 40000, 50000, 60000, 70000,
                         80000, 100000])

rho_w_25 = np.array([0.565, 0.475, 0.404, 0.345, 0.300,
                     0.205, 0.160, 0.130, 0.110, 0.098,
                     0.087, 0.072])   # ohm m

# Archie parameters
phi = 0.2
a = 0.62
m = 2
S = 1   # fully saturated
n = 2   # doesn't matter here because S=1

# Archie relation
rho_sandstone = a * (phi ** (-m)) * (S ** (-n)) * rho_w_25

# Print first question values
print("Sandstone resistivity at 25°C:")
for s, rw, r in zip(salinity_ppm, rho_w_25, rho_sandstone):
    print(f"Salinity = {s:6d} ppm | rho_w = {rw:.3f} ohm m | rho = {r:.3f} ohm m")

# Plot
plt.figure(figsize=(8, 5))
plt.plot(salinity_ppm, rho_w_25, marker='o', label=r'Pore water resistivity $\rho_w$ at 25°C')
plt.plot(salinity_ppm, rho_sandstone, marker='s', label=r'Sandstone resistivity $\rho$ at 25°C')

plt.xlabel('Salinity (ppm)')
plt.ylabel('Resistivity (ohm m)')
plt.title('Resistivity vs Salinity at 25°C')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
    
```
